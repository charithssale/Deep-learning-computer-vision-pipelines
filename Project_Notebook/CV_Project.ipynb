##1.	Implement any DL-based image classification algorithm except AlexNet


```python
import tensorflow as tf
from tensorflow.keras import layers, models
import matplotlib.pyplot as plt

# Load CIFAR-10 dataset
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.cifar10.load_data()

# Normalize pixel values
x_train = x_train.astype('float32') / 255.0
x_test = x_test.astype('float32') / 255.0

# One-hot encode labels
y_train = tf.keras.utils.to_categorical(y_train, 10)
y_test = tf.keras.utils.to_categorical(y_test, 10)

# Define a simple CNN (VGG-like architecture)
model = models.Sequential([
    layers.Conv2D(32, (3,3), activation='relu', input_shape=(32,32,3)),
    layers.Conv2D(32, (3,3), activation='relu'),
    layers.MaxPooling2D((2,2)),
    layers.Dropout(0.25),

    layers.Conv2D(64, (3,3), activation='relu'),
    layers.Conv2D(64, (3,3), activation='relu'),
    layers.MaxPooling2D((2,2)),
    layers.Dropout(0.25),

    layers.Flatten(),
    layers.Dense(512, activation='relu'),
    layers.Dropout(0.5),
    layers.Dense(10, activation='softmax')
])

model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# Train the model
history = model.fit(
    x_train, y_train,
    batch_size=64,
    epochs=10,
    validation_split=0.2
)

# Plot training and validation loss curves
plt.figure()
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Training and Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

## 2.	Implement any DL-based object detection (R-CNN, YOLO etc.) pipeline. You do not need to train the network. Run the inference using pretrained object detection models. Provide the object detection results for 5 different classes/objects.

```python
import os
import random
import sys
from PIL import Image, ImageDraw, ImageFont, UnidentifiedImageError
import torch
from torchvision.datasets import VOCDetection
from torchvision import transforms as T
from torchvision.models.detection import (
    fasterrcnn_resnet50_fpn,
    FasterRCNN_ResNet50_FPN_Weights
)
import matplotlib.pyplot as plt

# ─── CONFIG ───────────────────────────────────────────────
IMG_DIR     = "images"      # will be created; 5 VOC images placed here
NUM_SAMPLES = 5             # how many to sample/run
CONF_THRESH = 0.5           # detection confidence cutoff
os.makedirs(IMG_DIR, exist_ok=True)

# ─── 1) DOWNLOAD & SAMPLE ─────────────────────────────────
dataset = VOCDetection(root='.', year='2007', image_set='val', download=True)
indices = random.sample(range(len(dataset)), NUM_SAMPLES)
samples = []
for idx in indices:
    img, target = dataset[idx]
    fname = f"{idx}.jpg"
    path  = os.path.join(IMG_DIR, fname)
    img.save(path)
    objs = target['annotation']['object']
    if isinstance(objs, dict):
        objs = [objs]
    samples.append((path, objs))

# ─── 2) LOAD MODEL ────────────────────────────────────────
device  = torch.device("cuda" if torch.cuda.is_available() else "cpu")
weights = FasterRCNN_ResNet50_FPN_Weights.DEFAULT
model   = fasterrcnn_resnet50_fpn(weights=weights).to(device).eval()
transform       = T.Compose([T.ToTensor()])
COCO_CATEGORIES = weights.meta["categories"]

# VOC→COCO mapping (20 VOC classes)
voc_to_coco = {
    'aeroplane':'airplane','bicycle':'bicycle','bird':'bird','boat':'boat',
    'bottle':'bottle','bus':'bus','car':'car','cat':'cat','chair':'chair',
    'cow':'cow','diningtable':'dining table','dog':'dog','horse':'horse',
    'motorbike':'motorcycle','person':'person','pottedplant':'potted plant',
    'sheep':'sheep','sofa':'couch','train':'train','tvmonitor':'tv'
}

# ─── 3) IoU FUNCTION ───────────────────────────────────────
def iou(boxA, boxB):
    xA, yA = max(boxA[0], boxB[0]), max(boxA[1], boxB[1])
    xB, yB = min(boxA[2], boxB[2]), min(boxA[3], boxB[3])
    interW, interH = max(0, xB-xA), max(0, yB-yA)
    inter = interW * interH
    areaA = (boxA[2]-boxA[0])*(boxA[3]-boxA[1])
    areaB = (boxB[2]-boxB[0])*(boxB[3]-boxB[1])
    return inter / (areaA + areaB - inter + 1e-6)

# ─── 4) INFERENCE + EVALUATION ────────────────────────────
all_tp = all_fp = all_gt = 0

for img_path, gt_objs in samples:
    # Load & predict
    img = Image.open(img_path).convert("RGB")
    tensor = transform(img).to(device)
    with torch.no_grad():
        out = model([tensor])[0]

    # Parse predictions
    preds = []
    for box, lbl, scr in zip(out['boxes'], out['labels'], out['scores']):
        if scr < CONF_THRESH: continue
        preds.append((COCO_CATEGORIES[lbl], box.cpu().tolist(), float(scr)))

    # Parse ground truths
    gts = []
    for obj in gt_objs:
        voc = obj['name']
        if voc not in voc_to_coco: continue
        coco = voc_to_coco[voc]
        bb = obj['bndbox']
        gts.append((coco, [
            int(bb['xmin']), int(bb['ymin']),
            int(bb['xmax']), int(bb['ymax'])
        ]))

    # --- PRINT OUT PREDICTIONS & GROUND TRUTH ---
    print(f"\nImage: {os.path.basename(img_path)}")
    print(" Ground Truth Classes:")
    for coco, _ in gts:
        print(f"  - {coco}")
    print(" Predictions (class: score):")
    for coco, _, scr in preds:
        print(f"  - {coco}: {scr:.2f}")

    # Match GT→pred for metrics
    matched = set()
    tp = 0
    for coco, gt_box in gts:
        all_gt += 1
        best_i, best_val = -1, 0.0
        for i, (pname, pbox, _) in enumerate(preds):
            if i in matched or pname != coco: continue
            val = iou(gt_box, pbox)
            if val > best_val:
                best_val, best_i = val, i
        if best_val >= 0.5:
            tp += 1
            matched.add(best_i)
    all_tp += tp
    all_fp += len(preds) - len(matched)

    # Draw GT & predictions
    draw = ImageDraw.Draw(img)
    font = ImageFont.load_default()
    for coco, box in gts:
        draw.rectangle(box, outline="green", width=2)
        draw.text((box[0], box[1]-10), coco, fill="green", font=font)
    for coco, box, scr in preds:
        draw.rectangle(box, outline="red", width=2)
        draw.text((box[0], box[1]-10), f"{coco}:{scr:.2f}", fill="red", font=font)

    # Display
    plt.figure(figsize=(6,4))
    plt.imshow(img)
    plt.axis("off")
    plt.title(os.path.basename(img_path))
    plt.show()

# ─── 5) PRINT OVERALL METRICS ─────────────────────────────
precision = all_tp / (all_tp + all_fp + 1e-6)
recall    = all_tp / (all_gt + 1e-6)
print(f"\nOverall Precision: {precision:.3f}, Recall: {recall:.3f}\n")
```

## 3.	Implement vision transformer-based classification and compare the accuracy with results from Q1

```python
import torch, timm
from torch import nn, optim
from torch.cuda.amp import GradScaler, autocast
from torch.utils.data import DataLoader, Subset
import torchvision
from torchvision import transforms
from tqdm import tqdm

# ─── 0) QUICK SANITY PRINT ─────────────────────────────────────────────
print("→ Setting up data & model…")

# ─── 1) DATA: CIFAR-10, resized to 224×224, subset to 10k samples ─────
transform = transforms.Compose([
    transforms.Resize((224,224)),
    transforms.ToTensor(),
    transforms.Normalize((0.5,)*3, (0.5,)*3),
])
full_train = torchvision.datasets.CIFAR10('.', train=True,  download=True, transform=transform)
test_ds     = torchvision.datasets.CIFAR10('.', train=False, download=True, transform=transform)

# Use only first 10k for super-fast prototyping
train_ds    = Subset(full_train, list(range(10000)))

train_loader = DataLoader(train_ds,  batch_size=256, shuffle=True,  num_workers=0, pin_memory=False)
test_loader  = DataLoader(test_ds,   batch_size=256, shuffle=False, num_workers=0, pin_memory=False)

# ─── 2) MODEL: tiny distilled DeiT ────────────────────────────────────
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model = timm.create_model(
    'deit_tiny_distilled_patch16_224',
    pretrained=True,
    num_classes=10
).to(device)

# Freeze everything except the classification heads
for nm, p in model.named_parameters():
    if 'head' not in nm and 'distillation_head' not in nm:
        p.requires_grad = False

# ─── 3) LOSS & OPT ────────────────────────────────────────────────────
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(filter(lambda p: p.requires_grad, model.parameters()), lr=3e-4)
scaler    = GradScaler()

# ─── 4) TRAIN HEAD ONLY ────────────────────────────────────────────────
print("→ Starting head-only training on 10k samples…")
for epoch in range(1, 6):
    model.train()
    running = 0.0
    for imgs, labels in tqdm(train_loader, desc=f"Epoch {epoch}/5"):
        imgs, labels = imgs.to(device), labels.to(device)
        optimizer.zero_grad()
        with autocast():
            logits = model(imgs)
            loss   = criterion(logits, labels)
        scaler.scale(loss).backward()
        scaler.step(optimizer)
        scaler.update()
        running += loss.item() * imgs.size(0)
    print(f" Epoch {epoch} train loss: {running/len(train_ds):.4f}")

# ─── 5) EVALUATION ────────────────────────────────────────────────────
model.eval()
correct = 0
total   = 0
with torch.no_grad():
    for imgs, labels in test_loader:
        imgs, labels = imgs.to(device), labels.to(device)
        preds = model(imgs).argmax(dim=1)
        correct += (preds == labels).sum().item()
        total   += labels.size(0)

acc = correct/total
print(f"\nTiny-DeiT linear-probe (5 epochs, 10k train) Accuracy: {acc*100:.2f}%")
```

## 4.	Implement UNET or SegNet segmentation algorithm on any open source 2D/3D image segmentation dataset and provide dice coefficient.

```python
import os, random
import numpy as np
from PIL import Image
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader, Subset
from torchvision.datasets import VOCSegmentation
from torchvision import transforms
import torch.nn.functional as F

# ─── 1) HYPERPARAMETERS ────────────────────────────────────────────────
IMG_SIZE   = 128
BATCH_SIZE = 4
NUM_TRAIN  = 200    # small subset for speed
NUM_VAL    = 50
EPOCHS     = 10
LR         = 1e-3
DEVICE     = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

# ─── 2) CUSTOM VOC‐BINARY DATASET ──────────────────────────────────────
class VOCBinarySeg(Dataset):
    def __init__(self, root, year, image_set, download, img_size):
        self.voc = VOCSegmentation(
            root=root, year=year, image_set=image_set, download=download,
            transform=None, target_transform=None
        )
        self.img_size = img_size
        # transforms
        self.tf_img  = transforms.Compose([
            transforms.Resize((img_size, img_size)),
            transforms.ToTensor(),
        ])
        self.tf_mask = transforms.Compose([
            transforms.Resize((img_size, img_size)),
            transforms.ToTensor(),   # produces a float tensor [1,H,W]
        ])

    def __len__(self):
        return len(self.voc)

    def __getitem__(self, idx):
        img, mask = self.voc[idx]                # both are PIL Images
        # binarize the mask
        mask_np = np.array(mask)                 # H×W, values 0–255 (palette)
        bin_np  = (mask_np > 0).astype(np.uint8) # 1 where object, 0 background
        mask_bi = Image.fromarray(bin_np)        # PIL mode "L"
        # apply transforms
        img_t    = self.tf_img(img)              # 3×H×W
        mask_t   = self.tf_mask(mask_bi)         # 1×H×W, values 0.0 or 1.0
        return img_t, mask_t

# instantiate & subset
full_dataset = VOCBinarySeg(root='.', year='2012', image_set='train', download=True, img_size=IMG_SIZE)
indices = list(range(len(full_dataset)))
random.shuffle(indices)
train_idx, val_idx = indices[:NUM_TRAIN], indices[NUM_TRAIN:NUM_TRAIN+NUM_VAL]
train_dl = DataLoader(Subset(full_dataset, train_idx), batch_size=BATCH_SIZE, shuffle=True)
val_dl   = DataLoader(Subset(full_dataset, val_idx),   batch_size=BATCH_SIZE)

# ─── 3) UNet MODEL ─────────────────────────────────────────────────────
class UNet(nn.Module):
    def __init__(self):
        super().__init__()
        def C(in_ch, out_ch):
            return nn.Sequential(
                nn.Conv2d(in_ch, out_ch, 3, padding=1), nn.ReLU(inplace=True),
                nn.Conv2d(out_ch, out_ch, 3, padding=1), nn.ReLU(inplace=True),
            )
        self.d1 = C(3,64);   self.p1 = nn.MaxPool2d(2)
        self.d2 = C(64,128); self.p2 = nn.MaxPool2d(2)
        self.d3 = C(128,256);self.p3 = nn.MaxPool2d(2)
        self.b  = C(256,512)
        self.u3 = nn.ConvTranspose2d(512,256,2,2); self.up3 = C(512,256)
        self.u2 = nn.ConvTranspose2d(256,128,2,2); self.up2 = C(256,128)
        self.u1 = nn.ConvTranspose2d(128,64,2,2);  self.up1 = C(128,64)
        self.out= nn.Conv2d(64,1,1)

    def forward(self, x):
        d1, p1 = self.d1(x), self.p1(self.d1(x))
        d2, p2 = self.d2(p1), self.p2(self.d2(p1))
        d3, p3 = self.d3(p2), self.p3(self.d3(p2))
        b      = self.b(p3)
        u3     = self.up3(torch.cat([self.u3(b), d3], dim=1))
        u2     = self.up2(torch.cat([self.u2(u3), d2], dim=1))
        u1     = self.up1(torch.cat([self.u1(u2), d1], dim=1))
        return torch.sigmoid(self.out(u1))

# ─── 4) DICE COEFFICIENT ───────────────────────────────────────────────
def dice_coef(pred, target, eps=1e-6):
    # both are float tensors {0,1} shape (1,H,W)
    pred = pred.view(-1); target = target.view(-1)
    inter = (pred * target).sum()
    return (2*inter + eps) / (pred.sum() + target.sum() + eps)

# ─── 5) TRAIN & VALIDATE ───────────────────────────────────────────────
model = UNet().to(DEVICE)
opt   = optim.Adam(model.parameters(), lr=LR)
bce   = nn.BCELoss()

for ep in range(1, EPOCHS+1):
    model.train()
    running = 0.0
    for imgs, masks in train_dl:
        imgs, masks = imgs.to(DEVICE), masks.to(DEVICE)
        opt.zero_grad()
        out   = model(imgs)
        loss  = bce(out, masks)
        loss.backward()
        opt.step()
        running += loss.item()*imgs.size(0)
    print(f"Epoch {ep}/{EPOCHS} — Train BCE Loss: {running/len(train_idx):.4f}")

# compute mean Dice on validation subset
model.eval()
dices = []
with torch.no_grad():
    for imgs, masks in val_dl:
        imgs, masks = imgs.to(DEVICE), masks.to(DEVICE)
        preds = (model(imgs) > 0.5).float()
        for p, t in zip(preds, masks):
            dices.append(dice_coef(p, t).item())

mean_dice = sum(dices) / len(dices)
print(f"\nMean Dice Coefficient on {NUM_VAL} validation images: {mean_dice:.4f}")
```

## 5.	Implement GAN (any kind) and show at least 5 generated images.



```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader
import torchvision
from torchvision import transforms
import matplotlib.pyplot as plt

# 1. Hyperparameters
batch_size = 128
latent_dim = 100
epochs = 10
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

# 2. Data: MNIST normalized to [-1,1]
transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.5,), (0.5,))
])
train_ds = torchvision.datasets.MNIST(root='./data', train=True, download=True, transform=transform)
train_loader = DataLoader(train_ds, batch_size=batch_size, shuffle=True)

# 3. Generator
class Generator(nn.Module):
    def __init__(self):
        super().__init__()
        self.model = nn.Sequential(
            nn.Linear(latent_dim, 128*7*7),
            nn.BatchNorm1d(128*7*7),
            nn.ReLU(True),
            nn.Unflatten(1, (128, 7, 7)),
            nn.ConvTranspose2d(128, 64, 4, 2, 1),
            nn.BatchNorm2d(64),
            nn.ReLU(True),
            nn.ConvTranspose2d(64, 1, 4, 2, 1),
            nn.Tanh()
        )
    def forward(self, z):
        return self.model(z)

# 4. Discriminator
class Discriminator(nn.Module):
    def __init__(self):
        super().__init__()
        self.model = nn.Sequential(
            nn.Conv2d(1, 64, 4, 2, 1),
            nn.LeakyReLU(0.2, inplace=True),
            nn.Conv2d(64, 128, 4, 2, 1),
            nn.BatchNorm2d(128),
            nn.LeakyReLU(0.2, inplace=True),
            nn.Flatten(),
            nn.Linear(128*7*7, 1),
            nn.Sigmoid()
        )
    def forward(self, x):
        return self.model(x)

# 5. Initialize models
G = Generator().to(device)
D = Discriminator().to(device)
criterion = nn.BCELoss()
opt_G = optim.Adam(G.parameters(), lr=2e-4, betas=(0.5, 0.999))
opt_D = optim.Adam(D.parameters(), lr=2e-4, betas=(0.5, 0.999))

# 6. Training loop
for epoch in range(1, epochs+1):
    for real, _ in train_loader:
        real = real.to(device)
        b_size = real.size(0)
        # Labels
        real_labels = torch.ones(b_size, 1, device=device)
        fake_labels = torch.zeros(b_size, 1, device=device)
        # Train Discriminator
        z = torch.randn(b_size, latent_dim, device=device)
        fake = G(z)
        D_real = D(real)
        D_fake = D(fake.detach())
        loss_D = criterion(D_real, real_labels) + criterion(D_fake, fake_labels)
        opt_D.zero_grad()
        loss_D.backward()
        opt_D.step()
        # Train Generator
        D_fake2 = D(fake)
        loss_G = criterion(D_fake2, real_labels)
        opt_G.zero_grad()
        loss_G.backward()
        opt_G.step()
    print(f"Epoch {epoch}/{epochs} - Loss D: {loss_D:.4f}, Loss G: {loss_G:.4f}")

# 7. Generate and display 5 images
G.eval()
with torch.no_grad():
    z = torch.randn(5, latent_dim, device=device)
    samples = G(z).cpu()  # 5×1×28×28

fig, axes = plt.subplots(1, 5, figsize=(10, 2))
for i, img in enumerate(samples):
    img = (img * 0.5 + 0.5).squeeze()  # denormalize to [0,1]
    axes[i].imshow(img, cmap='gray')
    axes[i].axis('off')
plt.show()
```
